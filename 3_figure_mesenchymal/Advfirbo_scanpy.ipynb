```python
import numpy as np
import pandas as pd
import scanpy as sc
import scrublet as scr
import anndata
import scanpy.external as sce
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
mes = sc.read("./processed_data_py/COPD_mes.h5ad")
advfibro = mes[mes.obs["celltype"].isin(["Adventitial fibroblast"]) ].copy()
```

```python
sc.tl.pca(advfibro, n_comps = 50)
sc.external.pp.bbknn(advfibro, batch_key = "sample")
sc.tl.umap(advfibro)
sc.tl.leiden(advfibro,resolution = 0.5)
sc.pl.umap(advfibro, color=["leiden"],size=2, legend_loc='on data')
```

```python
advfibro.obs["cellsubtype"] = "Advfibro_1"
advfibro.obs.loc[advfibro.obs["leiden"].isin(["1"]),"cellsubtype"] = "Advfibro_2"
advfibro.obs.loc[advfibro.obs["leiden"].isin(["2"]),"cellsubtype"] = "Advfibro_3"
sc.pl.umap(advfibro, color=["cellsubtype"],size=2, legend_loc='on data')
```

```python
advfibro.obs['UMAP1'] = advfibro.obsm['X_umap'][:, 0]
advfibro.obs['UMAP2'] = advfibro.obsm['X_umap'][:, 1]
advfibro.obs.to_csv("./tables/advfibro_leiden_meta.csv")
advfibro.write("./processed_data_py/COPD_advfibro.h5ad")
```
