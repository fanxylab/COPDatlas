```python
import sys
import scanpy as sc
import scvelo as scv
import numpy as np
import anndata
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
mes = sc.read("./processed_data_py/COPD_advfibro.h5ad")
ldata = sc.read("processed_data_py/COPD_spilced.h5ad")
ldata = ldata[ldata.obs_names.isin(mes.obs_names)].copy()
```

```python
adata = scv.utils.merge(mes, ldata)
scv.pp.filter_and_normalize(adata, min_shared_counts=20, n_top_genes=2000)
scv.pp.moments(adata, n_pcs=50, n_neighbors=30)
scv.tl.recover_dynamics(adata)
scv.tl.velocity(adata, mode='dynamical')
scv.tl.velocity_graph(adata,n_jobs=10)
```

```python
scv.pl.velocity_embedding_stream(adata, basis='umap',color="cellsubtype",
                                 palette = ['#2ca02c', '#d62728', '#9467bd'],
                                 linewidth = 0.6, arrow_size= 0.8)
```
